```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json
```

```python
df = pd.read_json('new_emails.json')
```

```python
df.shape
```

```python
df['responded'] = df['responses'].apply(lambda x: True if len(x) > 0 else False)
```

```python
df['email_length'] = df['body'].apply(len)
df['email_length'].describe()

sns.boxplot(x='responded', y='email_length', data=df)
plt.title('Email Length vs. Response')
plt.savefig('email_length_vs_response.png')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
df['is_reply'].value_counts().plot.pie(autopct='%1.1f%%', labels=None, colors=['skyblue', 'lightgreen'])

plt.title('Distribution of responses', fontsize=16, color='white')
plt.ylabel('')
plt.legend(labels=['Not Replied', 'Replied'], loc='best', fontsize=12, facecolor='white') 
plt.tick_params(axis='x', colors='white') 

plt.setp(plt.gca().get_xticklabels(), fontsize=14, color='white')
plt.title('Distribution of responses', fontsize=16, color='black')
plt.savefig('response_distribution.png')
plt.show()
```

```python
df['day_of_week'] = df['date'].dt.day_name()

day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

df['day_of_week'].value_counts().loc[day_order].plot(kind='bar')

plt.title('Distribution of emails by day of the week')
plt.xlabel('Day of the week')
plt.ylabel('Number of emails')
plt.savefig('day_of_week.png')
plt.show()
```
