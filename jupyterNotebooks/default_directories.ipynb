```python
import os
```

```python
def find_common_folders(root_dir):
    folders = dict()

    for person in os.listdir(root_dir):
        person_path = os.path.join(root_dir, person)

        if os.path.isdir(person_path):
            for folder in os.listdir(person_path):
                folder_path = os.path.join(person_path, folder)
                
                if os.path.isdir(folder_path):
                    if folder in folders:
                        folders[folder] += 1
                    else:
                        folders[folder] = 1

                
    return folders
                        
```

```python
working_directory = os.getcwd() + '/../maildir'
folders = find_common_folders(working_directory)

folders = sorted(folders.items(), key=lambda x: x[1], reverse=True)

print(folders)
```
